```python
PDF_DIRECTORY = "/home/scott/projects/RedTideResearch/dataHarvests"
pdf_files = glob.glob("%s/*.pdf" % PDF_DIRECTORY)
```

```python
print(*pdf_files, sep='\n')
```

```python
#This works for single pages from a pdf 
for f in pdf_files:
    print("..........This is for a new pdf...............................")
    print(f)
    reader = PyPDF2.PdfFileReader(f)
    page = reader.getPage(4)
    print(page.extractText())
```
